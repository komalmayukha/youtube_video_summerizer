```python
prompt=f'"Answers questions based on:\ntext="{transcript_joined}.What does he talk about?"'
```

```python
from openai import OpenAI

# Initialize the Together.ai client (note: it's using OpenAI-compatible client)
client = OpenAI(
    api_key="f0af3fb0dfe6c1ff47bfcdf2109968303ee5f2c9556f9dad059204291df6e3a2",                  # Replace with your real key
    base_url="https://api.together.xyz/v1"            # Important: custom base URL
)

# Create a chat completion
chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": prompt,
        }
    ],
    model="mistralai/Mixtral-8x7B-Instruct-v0.1",     # Together.ai model
    temperature=1,
    max_tokens=256,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0
)

# Print the result
print(chat_completion.choices[0].message.content)
```
